```python
tracker_parsed = procurement_tracker.copy()
```

```python
tracker_parsed = tracker_parsed.rename(columns={"ALEMBA ": "ALEMBA"})
tracker_parsed["parsed_Request_Date"] = pd.to_datetime(tracker_parsed["Request_Date"], format="%d/%m/%Y", dayfirst=True, errors="coerce")
tracker_parsed["parsed_QTY"] = pd.to_numeric(tracker_parsed["QTY"], errors="coerce").astype("Int64")
tracker_parsed["parsed_Cost_PerUnit"] = pd.to_numeric(tracker_parsed["Cost_PerUnit"], errors="coerce")
tracker_parsed["parsed_Total_Cost"] = pd.to_numeric(tracker_parsed["Total_Cost"], errors="coerce")
tracker_parsed["parsed_Collection_Date"] = pd.to_datetime(tracker_parsed["Collection_Date"], format="%d/%m/%Y", dayfirst=True, errors="coerce")
tracker_parsed["parsed_ALEMBA"] = pd.to_numeric(tracker_parsed["ALEMBA"], errors="coerce")
tracker_parsed["parsed_ALEMBA"] = pd.to_numeric(tracker_parsed["ALEMBA"], errors="coerce").astype("Int64")
```

```python
import datetime

mandatory_fields = ["parsed_Request_Date", "FY", "parsed_ALEMBA", "Budget_Code", "Site", "Service_User", "Type", "Catalogue_Item", "Cost_PerUnit", "QTY", "Status", "Total_Cost", "Requisition_No", "BPS_Analyst"]
source_system = "TRACKER"
data_period = "23-24 to 24-25"
date_checked = datetime.date.today()


allowed_FY = ["23-24", "24-25"]
allowed_start_date = pd.to_datetime("2023-01-01")
allowed_end_date = pd.to_datetime("2026-03-31")
valid_sites = ["RLH", "CW", "SBH", "NUH", "MEH"]
```

```python
# costing calculations
tracker_parsed["expected_total"] = tracker_parsed["parsed_QTY"] * tracker_parsed["parsed_Cost_PerUnit"]


cost_ok = (
    tracker_parsed["parsed_QTY"].notna() &
    tracker_parsed["parsed_Cost_PerUnit"].notna() &
    tracker_parsed["parsed_Total_Cost"].notna()    
)


# per unit and total cost mismatch errors
tol = 0.01 # value point comparison
mismatch_calc = cost_ok & ((tracker_parsed["expected_total"] - tracker_parsed["parsed_Total_Cost"]).abs() > tol)
cost_mismatch_count = int(mismatch_calc.sum())


# if qty is 0 and total is not 0
# qty have value and total is 0
# if cost per unit is 0 but total and qty is not 0
```

```python
# missing field error
summary_records = []
for field in mandatory_fields:
    missing_count = tracker_parsed[field].isna().sum()
    summary_records.append({
        "error_type": "Missing Field",
        "source_system": source_system,
        "data_period": data_period,
        "date_checked": date_checked,
        "field_name": field,
        "error_count": missing_count
    })



# format error / invalid format
for field in mandatory_fields:
    raw_field = field.replace("parsed_", "")
    format_error_count = tracker_parsed[tracker_parsed[raw_field].notna() & tracker_parsed[field].isna()].shape[0]
    summary_records.append({
        "error_type": "Invalid Format",
        "source_system": source_system,
        "data_period": data_period,
        "date_checked": date_checked,
        "field_name": field,
        "error_count": format_error_count
    })



# out of range error - FY
out_of_range_FY_count = tracker_parsed[~tracker_parsed["FY"].isin(allowed_FY)].shape[0]
summary_records.append({
    "error_type": "Out of Range Values",
    "source_system": source_system,
    "data_period": data_period,
    "date_checked": date_checked,
    "field_name": "FY",
    "error_count": out_of_range_FY_count
})




# out of range error - date
out_of_range_date_count = tracker_parsed[tracker_parsed["parsed_Collection_Date"].notna() &
(
    (tracker_parsed["parsed_Collection_Date"] < allowed_start_date) |
    (tracker_parsed["parsed_Collection_Date"] > allowed_end_date)
)].shape[0]

summary_records.append({
    "error_type": "Out of Range Dates",
    "source_system": source_system,
    "data_period": data_period,
    "date_checked": date_checked,
    "field_name": "parsed_Collection_Date",
    "error_count": out_of_range_date_count
})



#out of range error - site
out_of_range_site_count = tracker_parsed[
    tracker_parsed["Site"].notna() &
    ~tracker_parsed["Site"].isin(valid_sites)].shape[0]

summary_records.append({
    "error_type": "Out of Range Values",
    "source_system": source_system,
    "data_period": data_period,
    "date_checked": date_checked,
    "field_name": "Site",
    "error_count": out_of_range_site_count
})


# per unit*QTY doesnt match the total cost column
summary_records.append({
    "error_type": "Cost Value Mismatch",
    "source_system": source_system,
    "data_period": data_period,
    "date_checked": date_checked,
    "field_name": "Total_Cost",
    "error_count": cost_mismatch_count
})
```

```python
error_summary = pd.DataFrame(summary_records)

error_summary["field_name"] = error_summary["field_name"].str.replace(r"^parsed_", "", regex=True)



field_names_report = {
    "Request_Date": "Request Date",
    "Budget_Code": "Budget Code",
    "Service_User": "Service User",
    "Catalogue_Item": "Catalogue Item",
    "Cost_PerUnit": "Cost Per Unit",
    "Total_Cost": "Total Cost",
    "Requisition_No": "Requisition Number",
    "BPS_Analyst": "BPS Analyst",
    "Collection_Date": "Collection Date"
}

error_summary["field_name"] = error_summary["field_name"].replace(field_names_report)
error_summary = error_summary[error_summary["error_count"] > 0]
```
