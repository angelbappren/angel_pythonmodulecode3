```python
# extract transformation

procurement_tracker = pd.read_excel(xlsx, sheet_name="Master", skiprows=4)
procurement_tracker = procurement_tracker.drop(columns=["Recharged", "RechgDate",
       "RechargedBy", "Funded", "Notes", "Requisition/PO",
       "Temp - Item details to be added in to Master Telephony Tracker",
       "Unnamed: 26", "Unnamed: 27", "Unnamed: 28", "Unnamed: 29"])
procurement_tracker = procurement_tracker.rename(columns={"`": "Request_Date", "BudgetCode": "Budget_Code", "BulkBucket": "Bulk_Bucket", "ISC ProjMgr": "ISC_ProjManager", "Service User": "Service_User", "CatalogueItem": "Catalogue_Item", "CollDate DD/MM/YYYY": "Collection_Date", "Requisition No:": "Requisition_No", "CostPerUnit": "Cost_PerUnit", "Total cost": "Total_Cost", "BPS Analyst ": "BPS_Analyst"})
```

```python
tracker_parsed = procurement_tracker.copy()
```

```python
tracker_parsed["parsed_Request_Date"] = pd.to_datetime(tracker_parsed["Request_Date"], format="%d/%m/%Y", dayfirst=True, errors="coerce")
tracker_parsed["parsed_QTY"] = pd.to_numeric(tracker_parsed["QTY"], errors="coerce").astype("Int64")
tracker_parsed["parsed_Total_Cost"] = pd.to_numeric(tracker_parsed["Total_Cost"], errors="coerce")
tracker_parsed["parsed_Collection_Date"] = pd.to_datetime(tracker_parsed["Collection_Date"], format="%d/%m/%Y", dayfirst=True, errors="coerce")
tracker_parsed["parsed_ALEMBA"] = pd.to_numeric(tracker_parsed["ALEMBA"], errors="coerce")
```

```python
tracker_parsed = tracker_parsed.rename(columns={"ALEMBA ": "ALEMBA"})
```

```python
tracker_parsed["parsed_ALEMBA"] = pd.to_numeric(tracker_parsed["ALEMBA"], errors="coerce").astype("Int64")
```

```python
# error - missing field draft
import datetime

mandatory_fields = ["parsed_Request_Date", "FY", "parsed_ALEMBA", "Budget_Code", "Site", "Service_User", "Type", "Catalogue_Item", "Cost_PerUnit", "QTY", "Status", "Total_Cost", "Requisition_No", "BPS_Analyst"]
source_system = "TRACKER"
data_period = "23-24 to 24-25"
date_checked = datetime.date.today()
```

```python
# missing field error
summary_records = []
for field in mandatory_fields:
    missing_count = tracker_parsed[field].isna().sum()
    summary_records.append({
        "error_type": "Missing Field",
        "source_system": source_system,
        "data_period": data_period,
        "date_checked": date_checked,
        "field_name": field,
        "count": missing_count
    })



# format error / invalid format
for field in mandatory_fields:
    raw_field = field.replace("parsed_", "")
    format_error_count = tracker_parsed[tracker_parsed[raw_field].notna() & tracker_parsed[field].isna()].shape[0]
    summary_records.append({
        "error_type": "Invalid format",
        "source_system": source_system,
        "data_period": data_period,
        "date_checked": date_checked,
        "field_name": field,
        "count": format_error_count
    })
```

```python
error_summary = pd.DataFrame(summary_records)

error_summary["field_name"] = error_summary["field_name"].str.replace(r"^parsed_", "", regex=True)



field_names_report = {
    "Request_Date": "Request Date",
    "Budget_Code": "Budget Code",
    "Service_User": "Service User",
    "Catalogue_Item": "Catalogue Item",
    "Cost_PerUnit": "Cost Per Unit",
    "Total_Cost": "Total Cost",
    "Requisition_No": "Requisition Number",
    "BPS_Analyst": "BPS Analyst"
}

error_summary["field_name"] = error_summary["field_name"].replace(field_names_report)
```
